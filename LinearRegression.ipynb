```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\Personal\ML\Machine Learning Practice\Salary_Data.csv")
df.head()
```

```python
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df["YearsExperience"], df["Salary"], color='blue' ,marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['YearsExperience']],df.Salary)
```

```python
predicted_salary = reg.predict([[1.1]])
print("Predicted Salary:", predicted_salary)
```

```python
plt.scatter(df["YearsExperience"], df["Salary"], color="blue")

plt.plot(df["YearsExperience"],
         reg.predict(df[["YearsExperience"]]),
         color="red")

plt.show()
```
